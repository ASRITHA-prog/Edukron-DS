```python
import numpy as np
import pandas as pd
data = {
    "Product": ["Laptop", "Mobile", "Tablet", "Headphones", "TV", "Fridge"],
    "Price": [80000, 25000, np.nan, 2000, 45000, np.nan],
    "Quantity": [5, np.nan, 15, 8, 7, 3],
    "Discount": [10, np.nan, 5, 2, np.nan, 8]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
df
```

```python
df.dropna(axis=0,how="any")
```

```python
df.dropna(axis=1,how="any")
```

```python
df.dropna(axis=0,how="all")
```

```python
df.dropna(axis=1,how="all")
```

```python
df = pd.DataFrame({
    "Customer": ["John", "Sara", np.nan, "Lily", "David", "Emma", "Chris", "Sophia"],
    "Age": [28, np.nan, np.nan, 40, np.nan, 30, 45, np.nan],
    "City": ["New York", "London", np.nan, "Sydney", "Paris", np.nan, "Berlin", "Tokyo"],
    "Purchase_Amount": [250, 400, np.nan, 150, 500, 600, np.nan, 350],
    "Feedback_Score": [5, np.nan, np.nan, 3, 5, np.nan, 2, 4]
})
df=pd.DataFrame(df)
df
```

```python
df.dropna(axis=0,how='all')
```

```python
df.dropna(axis=0,how='any')
```

```python
df.dropna(axis=1,how='all')
```

```python
df.dropna(axis=1,how='any')
```

```python
df['Age'].fillna(20)
```

```python
df['Age'].fillna(df['Age'].median(),inplace=True)
df
```
